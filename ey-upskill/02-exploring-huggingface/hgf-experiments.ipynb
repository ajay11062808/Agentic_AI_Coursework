### Hugginface - Basic Inference

```python
from transformers import pipeline
```

```python
model = pipeline("sentiment-analysis")
model
```

```python
model("Iam not sure if it is a right day for learning AI")
```

### Text Generation

```python
model = pipeline('text-generation', model="HuggingFaceTB/SmolLM2-360M-Instruct")
```

```python
text = model("Complete the sentence in 20 words, give a paragraph: gravity is a force that .... ")
```

```python
print(text[0]['generated_text'])
```

### Wrong use case

```python
model('Classify the statement into ["sports", "politics", "business", "education"]: This is an AI course')
```

### Classification 

```python
model = pipeline('zero-shot-classification', candidate_labels=["sports", "politics", "business", "education"])
```

```python
model('This is a course on Huggingface python module')
```
